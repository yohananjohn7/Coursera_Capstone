##### Remove the blank or NaN rows of CityMunRes

```python
df2 = df[df['CityMunRes'].notna()]
df2
```

```python
df_cases = df2[['CaseCode', 'ProvRes', 'CityMunRes']]
df_cases.head()
```

##### Rename the columns to match the columns of our coordinates file

```python
df_cases.rename(columns={'ProvRes':'Province', 'CityMunRes':'City'},inplace=True)
```

```python
df_cases.head()
```

#### Let us check which Province has the most high number of 

```python
df1 = df_cases.groupby('Province').count()

grouped = df1.groupby('Province').sum().reset_index()

grouped.sort_values('City', ascending=False)
```

#### So we are going to choose the Province of NCR

```python
df_ncr = df_cases[df_cases.Province == 'NCR']
df_ncr.head()
```

#### Remove the CITY OF and CITY string to match our coordinates file

```python
df_ncr['City'] = df_ncr['City'].replace({'CITY OF ':'', 'CITY':''}, regex=True)
#df_ncr['Province'] = df_ncr['Province'].replace({'NCR':'Metro Manila'}, regex=True)
```

```python
source = requests.get('https://github.com/zymartinez/philippines-lat-lng-list/blob/master/locations_latlng.csv').text
soup=BeautifulSoup(source,'lxml')
print(soup.title)
from IPython.display import display_html
tab = str(soup.table)
display_html(tab,raw=True)
```
